# 내장함수
사람들이 파이썬에 이미 만들어 놓은 함수
<pre>
ex) print()

### print()
값을 출력한다.

```
print('철수')
```

```
# ;를 쓰면 함수를 한 줄로 이어 쓸 수 있다.
print('철수');print('민수')
```

```
# ,end= 을 넣으면 뒤의 값이 따라오게 한다.
print('철수' ,end=' ')
print('민수')
```

```
print('철수', end=',');print('민수')
```

```
# 띄어 쓰기는 콤마로.
print('My', 'heart', 'will', 'go', 'on')
```

### type()
값의 타입을 알려준다.

```
a=1
type(a)
```

```
a='1'
type(a)
```

```
a=1.2
type(a)
```

```
a=[1,2,3]
type(a)
```

```
a=(1,2,3)
type(a)
```

```
a={'a':1, 'b':2}
type(a)
```

### len()
값의 길이(또는 값의 갯수)를 돌려줌.

```
a='황석현'
print(len(a))
```

```
a=[1,2,3,4]
print(len(a))
```

```
a=(1, 'z')
print(len(a))
```

### int()
'1' 같은 문자열 숫자나 1.2같은 소숫점 숫자를 정수로 돌려줌.

```
a='1'
b=int(a)
print(b)
```

```
type(b)
```

```
a=1.2
b=int(a)
print(b)
```

```
type(b)
```

### str()

```
a=1
b=str(1)
print(b)
```

```
type(b)
```

### list()
리스트로 만들어 돌려줌.

```
a='Julia'
b=list(a)
print(b)
```

```
a=list((1,2,3))
print(a)
```

### range()
입력받은 숫자의 범위 값을 돌려줌.

range(시작숫자, 끝숫자+1)

<pre>
사용법:
list(range())

```
a=list(range(0,5))
print(a)
```

```
# 시작 숫자는 생략 가능
a=list(range(5))
print(a)
```

```
a=list(range(5,10))
print(a)
```

```
### 숫자 사이 거리 추가
a=list(range(1,11,2))
print(a)
```

```
a=list(range(0,-11, -1))
print(a)
```
