# for문
횟수 혹은 값의 수 만큼 반복하는 제어문

## for문의 기본 구조
<pre>
* 형식:
  for 변수 in 리스트(또는 튜플, 문자열):
    수행할 문장1
    수행할 문장2
    ...

## for문 사용하기

```
list1 = ['a', 'b', 'c']

for i in list1:
  print(i)
```

```
a = [(2,4), (6,8), (9,10)]
for i, j in a:
  print(i, end=', ');print(j)
```

```
for i, j in a:
  print(i+j)
```

```
# 합/불합격 통보 프로그램
scores = [80, 70, 75, 85, 55]

num = 0
for score in scores:
  num = num+1
  if score >= 60:
    print('{}번 학생 합격'.format(num))
  else:
    print('{}번 학생 불합격'.format(num))
```

### enumerate()
순서가 있는 자료형(리스트, 튜플, 문자열)을 받아 인덱스(순번)과 값을 돌려줌.

for 문과 자주 사용됨.

```
for i, j in enumerate(['철수', '민수', '명수']):
  print('{}번 학생 이름은 {}입니다.'.format(i+1,j))
```

```
# 합/불합격 통보 프로그램
scores = [80,70,75,85,55]
for i, score in enumerate(scores):
  if score >= 60:
    print('{}번 학생 합격'.format(i+1))
  else:
    print('{}번 학생 불합격'.format(i+1))
```

## continue
for 문에 continue 이용

```
# 합격 통보기
scores = [80, 70, 75, 85, 55]

num =0
for score in scores:
  num = num + 1
  if score < 60:
    continue
  print('{}번 학생 합격입니다.'.format(num))
```

## range()
for 문에 range()이용

```
list1 = list(range(1,11))
print(list1)
```

```
# 1부터 10까지 더하기
sum = 0
for i in range(1, 11):
  sum = sum + i

print(sum)
```

```
# 리스트 길이를 모를 때 합격 통보기
scores = [90, 25, 67, 45, 80]
for i in range(len(scores)):
  if scores[i] < 60:
    continue
  print('{}번 학생 합격'.format(i+1))
```

## 구구단
for문을 중첩하여 구구단 만들기

```
for i in range(2,10):
  for j in range(1,10):
    print(i*j, end=' ')
  print(' ')
```

## 리스트 내포(List Comprehension)
편리 & 직관적
<pre>
* 사용법
리스트 = [(표현식) for i in (리스트) if (조건문)]

```
# 3곱하여 담기
a = [1,2,3,4]
r = []
for i in a:
  r.append(i*3)
  
print(r)
```

```
# 리스트 내포 사용
a = [1,2,3,4]
r = [i*3 for i in a]
print(r)
```

```
# 짝수에만 3 곱하여 담기
a = [1,2,3,4]
r = [i*3 for i in a if i%2 ==0]
print(r)
```

```
# 구구단
r = [i*j for i in range(2,10)
    for j in range(1,10)]

print(r)
```

```
# 홀수에만 2 곱하기
num = [1,2,3,4,5]
r = []
for i in num:
  if i%2 ==1:
    r.append(i*2)
  else:
    r.append(i)

print(r)
```

```
# list comprehension
num = [1,2,3,4,5]
r = [i*2 for i in num if i%2 ==1]

print(r)
```

```
# 학급 평균점수 구하기
A = [70, 60, 55, 80, 95, 100, 60, 85, 75, 80]
sum = 0

for i in A:
  sum = sum + i
  
mean = sum/len(A)

print(mean)
```

```
# 1~100까지 더하기
num = list(range(1,101))
sum = 0
for i in num:
  sum = sum + i
  
print(sum)
```

```
# 369게임
for i in range(1, 101):
  if i%3==0:
    print('짝')
  else:
    print(i)
```

```
# FizzBuzz
for i in range(1,101):
  if i %6 == 0:
    print('FizzBuzz')
  elif i %2 == 0:
    print('Fizz')
  elif i %3 == 0:
    print('Buzz')
  else:
    print(i)
```

```
# FizzBuzz
for i in range(1,101):
  if i %2 ==0 and i%3 ==0:
    print('FizzBuzz')
  elif i %2 == 0:
    print('Fizz')
  elif i %3 == 0:
    print('Buzz')
  else:
    print(i)
```
