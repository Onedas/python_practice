### x or y  
x, y 둘 중 하나만 참이면 참

### x and y 
x, y 둘 다 참이어야 참

### not x
x가 거짓이면 참

```
cash = 10000
card = True
```

```
if cash >= 5000 or card:
  print('식당에 간다.')
else:
  print('집에서 먹는다.')
```

```
# not을 쓰면 False로 만들어준다.
not cash
```

```
if not cash:
  print('집에서 먹는다')
else:
  print('외식 한다')
```

## 1-4.다양한 조건들은 elif
조건이 다양해지면 elif를 사용
<pre>
사용법:
if 조건문1:
  실행문1
elif 조건문2:
  실행문2
elif 조건문3:
  실행문3
else:
  실행문4

```
store = ['크림빵', '가위', '연필']
```

```
if '소보로' in store:
  print('소보로를 산다')
elif '크림빵' in store:
  print('크림빵을 산다.')
else:
  print('상점을 나온다.')
```

### if문 한 줄로 작성하기

```
store = ['크림빵', '가위', '연필']
```

```
if '크림빵' not in store:
  pass
else:
  print('상점을 나온다.')
```

```
# 한 줄로 작성
if '크림빵' not in store: pass
else: print('상점을 나온다.')
```

## 1-6. 조건부 표현식
<pre>
*사용법:

(조건문이 참일 때의 실행문) (if 조건문) (else) (조건문이 거짓일 때의 실행문) 

```
score = 91
```

```
# 1.출력만 하기
if score >= 70:
  print('pass')
else:
  print('fail')
```

```
# 1에 대한 조건부 표현식
print('pass') if score >= 70 else print('fail')
```

```
# 2.변수에 넣기
if score >= 70:
  my_score = 'pass'
else:
  my_score = 'fail'
```

```
print(my_score)
```

```
# 2에 대한 조건부 표현식
my_score = 'pass' if score >= 70 else 'fail'
```

```
print(my_score)
```
