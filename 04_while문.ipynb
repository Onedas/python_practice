# 2.While문
조건문이 성립하는 동안 실행문을 반복 실행

## 2-1.while문 구조
<pre>
while <조건문>:
  <실행문1>
  <실행문2>
  ..

```
# 'python' 10번 출력하는 프로그램 만들기
i = 0
while i<10:
  print('python')
  i=i+1
```

```
# '*'을 10번 출력하는 프로그램 만들기
i=0
while i<10:
  print('*')
  i=i+1
```

```
# '*'를 5번 출력하는 프로그램 만들기
i=0
while i<10:
  print('*')
  i=i+2
```

```
# '*'를 5번 출력하는 프로그램
i=5
while i>0:
  print('*')
  i=i-1
```

```
# '*'를 한개씩 증가시키면서 출력하라
i=0
while i<5:
  i=i+1
  print('*'*i)
```

```
# '*'를 1개씩 증가시키면서 출력하는 프로그램 만들기
i=10
while i>1:
  print('*'*i)
  i= i-1
```

```
# 열번 찍어 안넘어 가는 나무 없다
hit = 0
while hit < 10:
  hit = hit +1
  print('나무를 {}번 찍었다.'.format(hit))
  if hit == 10:
    print('나무가 넘어간다.')
```

```
# 10번 세워 안세워지는 나무 없다
hit = 10
while hit>0:
  print('{}번 세웠습니다'.format(hit))
  hit=hit-1
  if hit == 0:
    print('나무를 재건했습니다.')
```

### while문 강제 종료
break 사용
<pre>
자판기 콜라가 다 떨어졌을 때
판매 중지 메세지 주고
강제 종료

```
cola = 5
money = True

while money:
  print('콜라가 나옵니다.')
  cola = cola -1
  print('남은 콜라 양은 {}개 입니다.'.format(cola))
  if cola == 0:
    print('콜라가 다 떨어졌습니다. 판매를 중지 합니다.')
    break
```

```
cola = 5
money = 5000
while True:
  if money == 1000:
    print('콜라가 나옵니다')
    cola = cola -1
  elif money >1000:
    print('콜라가 나옵니다.')
    money = money-1000
    print('거스름돈으로 {}가 나옵니다.'.format(money))
    cola = cola -1
  else:
    print('돈이 부족합니다.')
  if cola ==0:
    print('재고가 없습니다.')
    print('판매를 중지합니다.')
    break
```

### 자판기 만들기

```
cola = 5
while True:
  money = int(input('돈을 넣어주세요.>'))
  if money == 1000:
    print('콜라가 나옵니다')
    cola = cola -1
    print('콜라 재고는 {}개 입니다.'.format(cola))
  elif money >1000:
    print('콜라가 나옵니다.')
    print('거스름돈으로 {}가 나옵니다.'.format(money-1000))
    cola = cola -1
    print('콜라 재고는 {}개 입니다.'.format(cola))
  else:
    print('돈이 부족합니다.')
  if cola ==0:
    print('재고가 없습니다.')
    print('판매를 중지합니다.')
    break
```

### break의 반대 : continue
break로 while 문을 끝 냈음.

continue는 처음으로 다시 돌아감.


```
a=0
while a<10:
  a += 1 # a= a+1과 같음
  if a % 2 == 0:
    continue
  print(a)
```
